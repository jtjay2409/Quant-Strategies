To understand both, you must first know that options traders look at an "Implied Volatility Smile"—a chart showing that options with different strike prices have different expected volatilities. Both models try to explain why this smile exists.

## • Local Volatility (The Dupire Model)

• The Concept: It says volatility is deterministic. This means if you tell the model the exact current stock price (S) and the exact time (t), it will spit out one exact volatility number.
• How it views the world: Volatility changes, but only because the stock price or time changed. It assumes there is a fixed map of volatility spread across different price levels.

## • Stochastic Volatility (The Heston Model / SABR Model)

• The Concept: It says volatility is random. Volatility is driven by its own separate roll of the dice, completely independent of whether the stock price moves up or down.
How it views the world: Two identical stocks could be sitting at exactly $100 today, but one could have massive volatility and the other could have very low volatility because their "volatility weather" is different.

### 2. The "Smile" Dynamics (Why it matters)
The biggest difference shows up when the stock price actually moves. Traders need to know how their volatility outlook will change tomorrow if the market crashes or rallies today.
Local Volatility (Predictable Shift)
Price Drops ──> Automatically moves to a higher vol zone on the "map".
Price Rises ──> Automatically moves to a lower vol zone on the "map".
Stochastic Volatility (Unpredictable Shift)

Price Drops ──> Volatility might skyrocket, or it might stay calm (random shock).
• Local Volatility's Flaw (The Forward Smile): Because it relies on a fixed map, if the stock price moves to a new level tomorrow, the model predicts the "smile" will shift in a highly unrealistic, rigid way. It often underestimates how volatile the future will actually be.
• Stochastic Volatility's Strength: It allows the smile to twist, flatten, or shift up and down randomly, which perfectly matches how real human fear and greed change in the stock market.

### 3. Hedging: How Traders Decide
Traders do not just buy options; they must "hedge" (manage the risk) of those options.
• Delta Hedging with Local Vol: Because Local Vol assumes a direct link between the stock price and volatility, it calculates a very specific "Delta" (how many shares to buy/sell to cover risk). However, because the real world is random, this hedge often performs poorly in market crashes.
• Vega Hedging with Stochastic Vol: Because Stochastic Vol treats volatility as a separate random variable, it forces the trader to hedge against changes in volatility itself (called Vega hedging). This requires buying other options to protect against a sudden spike in market fear.
